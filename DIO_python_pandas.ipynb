```python
## leitura arquivo ponto_3.csv

import pandas as pd
import numpy as np


ponto = pd.DataFrame(pd.read_csv("/content/ponto_3.csv", sep=",", encoding='utf-8')) 

# limpeza da planilha
# substituindo valores nulos na coluna Contrato pela matricula do funcionário
ponto['Contrato'].fillna(method='ffill', inplace=True)

ponto
```

```python
# eliminando linhas que indicam falta não justificada

ponto = ponto.dropna(subset=["Marcações Realizadas"])
```

```python
# eliminando celulas que indicam que o funcionario estava de folga

ponto = ponto.dropna(subset=["Hor"])
ponto

```

```python
# importando arquivo turnos

turnos = pd.DataFrame(pd.read_csv("/content/turnos3.csv", sep=";", encoding='utf-8')) 
turnos
```

```python
ponto['Equipe'] = 'NaN'

for i, row in ponto.iterrows():
  for j, r in turnos.iterrows():
    if row['Contrato'] == r['MATR.']:
      ponto.loc[i,'Equipe'] =  turnos.loc[j,'TURNO']
      break
ponto
  
```
